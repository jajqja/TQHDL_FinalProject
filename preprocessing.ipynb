Điền NaN

```python
raw_df['Frontage'] = raw_df['Frontage'].fillna(0.0)
```

```python
raw_df['Access Road'] = raw_df['Access Road'].fillna(0.0)
```

```python
raw_df['House direction'] = raw_df['House direction'].fillna('Không xác định')
```

```python
raw_df['Balcony direction'] = raw_df['Balcony direction'].fillna('Không xác định')
```

```python
raw_df['Furniture state'] = raw_df['Furniture state'].fillna('Empty')
```

```python
raw_df['Legal status'] = raw_df['Legal status'].fillna('Unknown')
```

```python
raw_df = raw_df.dropna() # còn lại NaN ở cột Bedrooms với Bathrooms tui xóa luôn. Tại vì 2 cột đố để 0 thì vô lý mà != 0 mình cũng không chắc chắn.
```

```python
raw_df.head(10)
```

```python
raw_df.isna().sum()
```

```python
raw_df = raw_df.reset_index(drop=True)
```

```python
raw_df['Province'] = raw_df['Address'].apply(lambda x: x.split(',')[-1].replace('.','').split('(')[0].strip())
index = []
for i in range(len(raw_df['Province'])):
    t = raw_df['Province'].iloc[i]
    if t == 'giá 6ty':
        raw_df['Province'][i] = 'Hồ Chí Minh'
    elif t in ['TP HCM', 'TpHCM', 'TPHCM', 'TP Hồ Chí Minh', 'Hồ Chí Mính', 'Quận 8', 'Quận Bình Thạnh']:
        raw_df['Province'][i] = 'Hồ Chí Minh'
    elif t in ['HN', 'Quận Nam Từ Liêm']:
        raw_df['Province'][i] = 'Hà Nội'
    elif t in ['Đường số 11', 'Phòng công chứng Nguyễn Thị Thành']:
        index.append(i)

raw_df = raw_df.drop(index)
raw_df['Province'].value_counts()
```

```python
raw_df['Project'] = raw_df['Address'].apply(lambda x: 'Yes' if x.split(',')[0].find('Dự án')!=-1 else 'No')
```
